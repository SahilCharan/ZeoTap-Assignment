```python
import pandas as pd

# Load the datasets
customers_df = pd.read_csv('Customers.csv')
transactions_df = pd.read_csv('Transactions.csv')
products_df = pd.read_csv('Products.csv')
```

```python
# Merge transactions with customers to associate customer profile information
merged_data = transactions_df.merge(customers_df, on="CustomerID", how="inner")

# Aggregating data to create customer-level metrics
customer_metrics = merged_data.groupby("CustomerID").agg(
    TotalSpending=("TotalValue", "sum"),
    TransactionCount=("TransactionID", "count"),
    AverageTransactionValue=("TotalValue", "mean"),
    FirstTransaction=("TransactionDate", "min"),
    LastTransaction=("TransactionDate", "max"),
    Region=("Region", "first"),
).reset_index()

# Convert transaction dates to datetime format for further analysis
customer_metrics["FirstTransaction"] = pd.to_datetime(customer_metrics["FirstTransaction"])
customer_metrics["LastTransaction"] = pd.to_datetime(customer_metrics["LastTransaction"])

# Calculate Recency (days since last transaction)
customer_metrics["Recency"] = (pd.Timestamp.now() - customer_metrics["LastTransaction"]).dt.days

# Drop unnecessary date columns for clustering
customer_metrics = customer_metrics.drop(columns=["FirstTransaction", "LastTransaction"])

# Display the processed customer metrics
customer_metrics.head()
```

```python


import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns


customers_df = pd.read_csv('Customers.csv')
transactions_df = pd.read_csv('Transactions.csv')

# Feature scaling
scaler = StandardScaler()
features_to_scale = ['TotalSpending', 'TransactionCount', 'AverageTransactionValue', 'Recency']
customer_metrics[features_to_scale] = scaler.fit_transform(customer_metrics[features_to_scale])

# Determine optimal number of clusters (e.g., using the elbow method or silhouette analysis)
# For demonstration, we'll use KMeans with k=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10) # Increased n_init for better convergence
customer_metrics['Cluster'] = kmeans.fit_predict(customer_metrics[features_to_scale])

# Evaluate the clustering using Davies-Bouldin Index
db_index = davies_bouldin_score(customer_metrics[features_to_scale], customer_metrics['Cluster'])
print(f"Davies-Bouldin Index: {db_index}")

# Visualize the clusters (example with two features)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='TotalSpending', y='Recency', hue='Cluster', data=customer_metrics, palette='viridis')
plt.title('Customer Segmentation')
plt.show()




print("Clustering Results:")
print(f"Number of Clusters: 3")
print(f"Davies-Bouldin Index: {db_index}")
```

```python
# Check the columns of the merged data to ensure 'Price' exists
merged_data = transactions_df.merge(customers_df, on='CustomerID')
merged_data = merged_data.merge(products_df, on='ProductID')

print("Columns in merged data:")
print(merged_data.columns)

# Display a sample of the merged data to verify
print("\
Sample of merged data:")
print(merged_data.head())
```

```python
# Build the lookalike model with corrected features
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
import numpy as np

# Prepare customer profiles
customer_profile = merged_data.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum',
    'Price_x': 'mean',
    'Region': 'first',
    'Category': lambda x: ','.join(set(x))
}).reset_index()

# One-hot encode categorical variables
customer_profile = pd.get_dummies(customer_profile, columns=['Region'])

# Scale numerical features
scaler = StandardScaler()
numerical_features = ['TotalValue', 'Quantity', 'Price_x']
customer_profile[numerical_features] = scaler.fit_transform(customer_profile[numerical_features])

# Compute similarity matrix
similarity_matrix = cosine_similarity(customer_profile.drop(columns=['CustomerID', 'Category']))

# Find top 3 similar customers for first 20 customers
lookalike_results = []
for i in range(20):
    customer_id = customer_profile['CustomerID'].iloc[i]
    similar_indices = np.argsort(-similarity_matrix[i])[1:4]
    similar_customers = customer_profile.iloc[similar_indices]['CustomerID'].values
    similarity_scores = similarity_matrix[i][similar_indices]

    for similar_customer, score in zip(similar_customers, similarity_scores):
        lookalike_results.append({
            'CustomerID': customer_id,
            'SimilarCustomerID': similar_customer,
            'SimilarityScore': round(score, 4)
        })

# Create and save results
lookalike_df = pd.DataFrame(lookalike_results)
lookalike_df.to_csv('Lookalike.csv', index=False)

print("Top 3 lookalikes for first 20 customers:")
print(lookalike_df.head(10))


# Perform clustering analysis using K-means
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns

# Select features for clustering
clustering_features = customer_profile.drop(columns=['CustomerID', 'Category'])

# Determine the optimal number of clusters (2 to 10) using Davies-Bouldin Index
cluster_range = range(2, 11)
db_scores = []

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(clustering_features)
    db_index = davies_bouldin_score(clustering_features, labels)
    db_scores.append(db_index)

# Plot Davies-Bouldin Index for each number of clusters
plt.figure(figsize=(10, 6))
sns.lineplot(x=list(cluster_range), y=db_scores, marker='o')
plt.title('Davies-Bouldin Index for Different Numbers of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Davies-Bouldin Index')
plt.xticks(cluster_range)
plt.grid()
plt.show()

# Choose the optimal number of clusters (minimum DB Index)
optimal_k = cluster_range[db_scores.index(min(db_scores))]
print("Optimal number of clusters based on Davies-Bouldin Index:", optimal_k)

# Perform clustering with the optimal number of clusters
final_kmeans = KMeans(n_clusters=optimal_k, random_state=42)
customer_profile['Cluster'] = final_kmeans.fit_predict(clustering_features)

# Save the clustered data
customer_profile.to_csv('Customer_Segmentation.csv', index=False)

print("Clustering completed and results saved to 'Customer_Segmentation.csv'.")
```

```python
# Create cluster visualization using PCA
from sklearn.decomposition import PCA

# Prepare data for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(clustering_features)

# Create scatter plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(pca_result[:, 0], pca_result[:, 1], c=customer_profile['Cluster'], cmap='tab10')
plt.title('Customer Segments Visualization (PCA)')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.colorbar(scatter, label='Cluster')
plt.show()

# Generate summary statistics for each cluster
cluster_summary = customer_profile.groupby('Cluster').agg({
    'TotalValue': ['mean', 'count'],
    'Quantity': 'mean',
    'Price_x': 'mean'
}).round(2)

print("\
Cluster Summary:")
print(cluster_summary)

# Save final report
report = f"""Customer Segmentation Analysis Report
----------------------------------------
Number of Clusters: {optimal_k}
Davies-Bouldin Index: {min(db_scores):.4f}

Key Findings:
- Optimal number of clusters determined using Davies-Bouldin Index: {optimal_k}
- Total customers segmented: {len(customer_profile)}
"""

with open('segmentation_report.txt', 'w') as f:
    f.write(report)

print("\
Full report saved to 'segmentation_report.txt'")
```

```python
# Try different clustering algorithms and evaluate them
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import numpy as np

# Prepare data
X = clustering_features.values

results = []
cluster_ranges = [2, 4, 6, 8, 10]  # Testing different numbers of clusters

# Test KMeans
for n_clusters in cluster_ranges:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(X)

    results.append({
        'Algorithm': 'KMeans',
        'n_clusters': n_clusters,
        'DB_Index': davies_bouldin_score(X, labels),
        'Silhouette': silhouette_score(X, labels),
        'CH_Score': calinski_harabasz_score(X, labels)
    })

# Test Hierarchical Clustering
for n_clusters in cluster_ranges:
    hc = AgglomerativeClustering(n_clusters=n_clusters)
    labels = hc.fit_predict(X)

    results.append({
        'Algorithm': 'Hierarchical',
        'n_clusters': n_clusters,
        'DB_Index': davies_bouldin_score(X, labels),
        'Silhouette': silhouette_score(X, labels),
        'CH_Score': calinski_harabasz_score(X, labels)
    })

# Test DBSCAN with different eps values
eps_values = [0.5, 1.0, 1.5]
for eps in eps_values:
    dbscan = DBSCAN(eps=eps)
    labels = dbscan.fit_predict(X)

    if len(np.unique(labels)) > 1:  # Only evaluate if more than one cluster is formed
        results.append({
            'Algorithm': f'DBSCAN (eps={eps})',
            'n_clusters': len(np.unique(labels[labels >= 0])),
            'DB_Index': davies_bouldin_score(X, labels),
            'Silhouette': silhouette_score(X, labels),
            'CH_Score': calinski_harabasz_score(X, labels)
        })

# Convert results to DataFrame and sort by DB Index
results_df = pd.DataFrame(results)
results_df = results_df.sort_values('DB_Index')

print("Clustering Algorithm Comparison (sorted by DB Index):")
print(results_df)

# Find best model based on DB Index
best_result = results_df.iloc[0]
print("\
Best clustering configuration:")
print(f"Algorithm: {best_result['Algorithm']}")
print(f"Number of clusters: {best_result['n_clusters']}")
print(f"DB Index: {best_result['DB_Index']:.4f}")
print(f"Silhouette Score: {best_result['Silhouette']:.4f}")
print(f"Calinski-Harabasz Score: {best_result['CH_Score']:.4f}")

# Save results
results_df.to_csv('clustering_comparison.csv', index=False)
```

To analyze specific cluster characteristics and generate additional metrics, I will calculate summary statistics for each cluster (e.g., mean values of key features) and evaluate additional clustering metrics like intra-cluster variance and inter-cluster distance. This will provide deeper insights into the clusters.

```python
# Analyze specific cluster characteristics
cluster_characteristics = customer_profile.groupby('Final_Cluster').agg({
    'TotalValue': ['mean', 'sum'],
    'Quantity': ['mean', 'sum'],
    'Price_x': 'mean'
}).round(2)

print("Cluster Characteristics:")
print(cluster_characteristics)

# Generate additional metrics: intra-cluster variance and inter-cluster distance
from sklearn.metrics import pairwise_distances

# Calculate intra-cluster variance
intra_cluster_variance = []
for cluster in range(10):
    cluster_points = X[labels == cluster]
    if len(cluster_points) > 1:
        variance = np.mean(pairwise_distances(cluster_points))
        intra_cluster_variance.append(variance)
    else:
        intra_cluster_variance.append(0)

# Calculate inter-cluster distance
centroids = best_kmeans.cluster_centers_
inter_cluster_distance = np.mean(pairwise_distances(centroids))

print("\
Additional Metrics:")
print("Intra-cluster Variance (per cluster):", intra_cluster_variance)
print("Inter-cluster Distance:", inter_cluster_distance)

# Save cluster characteristics and metrics
cluster_characteristics.to_csv('cluster_characteristics.csv')
with open('additional_metrics.txt', 'w') as f:
    f.write("Intra-cluster Variance (per cluster): " + str(intra_cluster_variance) + "\
")
    f.write("Inter-cluster Distance: " + str(inter_cluster_distance) + "\
")

print("Cluster characteristics and additional metrics saved.")
```
